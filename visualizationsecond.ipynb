```python
import matplotlib.pyplot as plt
```

```python
x = range(-10,10)
y = [x*x -x + 2 for x in x]
```

```python
plt.plot(x,y)
plt.annotate('Origin', xy=(0,0), xytext=(0,20), arrowprops=dict(arrowstyle='->'))
```

```python
import math
```

```python
x = range(0,20)
y = list(math.sqrt(n) for n in x)
```

```python
plt.plot(x,y,
        label = 'y=√x line',
        linestyle = '--')
plt.title('y=√x chart')
plt.xlabel('x range')
plt.ylabel('y range')
```

```python
x = range(0,20)
y = list(n**3 for n in x)
plt.plot(x,y,
        label = 'y=x**3 line',
        linestyle = '--')
plt.title('y=x**3chart')
plt.xlabel('x range')
plt.ylabel('y range')
```

```python
x = range(0,20)
y = list(3**n for n in x)
plt.plot(x,y,
        label = 'y=2**n line',
        linestyle = '--')
plt.title('y=2**n chart')
plt.xlabel('x range')
plt.ylabel('y range')
```

```python
x = range(0,20)
y = list(1/(n+1) for n in x)
plt.plot(x,y,
        label = 'y=1/(x+1) line',
        linestyle = '--')
plt.title('y=1/(x+1) chart')
plt.xlabel('x range')
plt.ylabel('y range')
```

```python
x = range(0,11)
y1 = list(math.sqrt(n) for n in x)
y2 = list(n**3 for n in x)
y3 = list(2**n for n in x)
y4 = list(1/(n+1) for n in x)

plt.subplot(2,2,1)
plt.plot(x,y1)

plt.subplot(2,2,2)
plt.plot(x,y2)

plt.subplot(2,2,3)
plt.plot(x,y3)

plt.subplot(2,2,4)
plt.plot(x,y4)
```
