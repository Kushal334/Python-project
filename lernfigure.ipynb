```python
import matplotlib 
import matplotlib.pyplot as plt
import numpy as np 
t = np.arange(0.0,4.0, 0.001)
s = 1+np.sin(2*np.pi *t)
```

```python
fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks')
ax.grid()

fig.savefig("test.png")
plt.show()
```
