```python
from matplotlib import pyplot as plt
fig =plt.figure()
ax= fig.add_axes([0,0,1,1])
x=[1,4,-1,0,-2]
y=[2,7,3,-1,-3]
colors= ['r', 'g', 'y','b','b']
shapes= [ 'o','*','s','^','^']
ax.set_xlabel('X AXIS')
ax.set_ylabel('Y AXIS')
#ax.scatter(x,y,color=colors, marker=shapes)
for i in range(len(x)):
  ax.scatter(x[i], y[i], color=colors[i], marker=shapes[i])
ax.set_title('Scatter Plot')
plt.show()
```

```python

fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
x= ["A","B","C","D","E","F"]
y= [7,3,14,4,2,1]
ax.set_xlabel("Categories")
ax.set_ylabel("Heights")
ax.bar(x,y)
plt.show()
```

```python
import numpy as np
fig= plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.set_title("The Graph of y = 2x + 3")
x = np.linspace(0,10,100)
y= 2*x+3
ax.plot(x,y , label="y = 2x + 3")
plt.show()
```
