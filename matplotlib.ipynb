```python
plt.figure(figsize = (16,16))
plt.title('Big O Notation')
plt.xlabel('Elements')
plt.ylabel('Operations')
x = range(1,10)
y1 = [num * 0 + 1 for num in x]
y2 = [math.log(num) for num in x]
y3 = [num for num in x]
y4 = [num * math.log(num) for num in x]
y5 = [num ** 2 for num in x]
y6 = [2 ** num for num in x]
y7 = [math.factorial(num) for num in x]
y8 = [num ** num for num in x]
plt.scatter(x, y1, color = 'r', label = ('O(1)'))
plt.plot(x,y1)
plt.scatter(x, y2, color = 'g', label = ('O(log n)'))
plt.plot(x,y2)
plt.scatter(x, y3, color = 'b', label = ('O(n)'))
plt.plot(x,y3)
plt.scatter(x, y4, color = 'm', label = ('O(n log n)'))
plt.plot(x,y4)
plt.scatter(x, y5, color = 'c', label = ('O($n^2$)'))
plt.plot(x,y5)
plt.scatter(x, y6, color = 'y', label = ('O($2^n$)'))
plt.plot(x,y6)
plt.scatter(x, y7, color = 'k', label = ('O(n!)'))
plt.plot(x,y7)
plt.scatter(x, y8, color = 'aquamarine', label = ('O($n^n$)'))
plt.plot(x,y8)

plt.ylim(0,100)

plt.legend()
```

```python
x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.figure(figsize = (14,14))

plt.subplot(2,2,1)
plt.plot(x,y1)
plt.title('x^(1/2)')

plt.subplot(2,2,2)
plt.plot(x,y2)
plt.title('$x^3$')

plt.subplot(2,2,3)
plt.plot(x,y3)
plt.title('$2^x$')

plt.subplot(2,2,4)
plt.plot(x,y4)
plt.title('1/(x+1)')
```

```python

x = range(10)
y1 = [num ** (1/2) for num in x]
y2 = [num ** 3 for num in x]
y3 = [2 ** num for num in x]
y4 = [1/(num + 1) for num in x]

plt.plot(x,y1)
plt.scatter(x,y1, color = 'r', label = 'x^(1/2)')
plt.plot(x,y2)
plt.scatter(x,y2, color = 'b', label = '$x^3$')
plt.plot(x,y3)
plt.scatter(x,y3, color = 'g', label = '$2^x$')
plt.plot(x,y4)
plt.scatter(x,y4, color = 'm', label = '1/(x+1)')
plt.title('Four Different Functions')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
