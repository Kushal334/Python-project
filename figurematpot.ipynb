```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import make_blobs
```

```python

X1, y = make_blobs(random_state=170, n_samples=600, centers = 5)
rng = np.random.RandomState(74)

transformation = rng.normal(size=(2, 2))
X = np.dot(X1, transformation)


plt.scatter(X[:, 0], X[:, 1])
plt.xlabel("Feature 0")
plt.ylabel("Feature 1")
plt.show()
```

```python
def show_model_prediction(model, X):
  labels = model.fit_predict(X)
  plt.scatter(X[:, 0], X[:, 1], c=labels,
            s=30, cmap='viridis');
```

```python
from sklearn.cluster import AgglomerativeClustering
# Instantiate and fit your model here
model_1 =AgglomerativeClustering(n_clusters=5)
model_1.fit(X)
show_model_prediction(model_1, X)
```

```python
from matplotlib import style
month = ["Jan","feb","Mar","April","may"]
holidays= [2,4,5,2,9]
IMP_Tasks = [1,2,4,3,6]

style.use("ggplot")

plt.plot(month,holidays,"mo--", linewidth=3,markersize=10 , label="Holidays")
plt.plot(month,IMP_Tasks,"go:", linewidth=3,markersize=10 ,label="Tasks")

plt.axis([0,"may", 0,9]) 

plt.title("No of holidays and tasks in a month", fontsize = 15)
plt.xlabel("Months", fontsize = 13)
plt.ylabel("Holidays", fontsize = 13)

plt.legend(loc=4)

plt.grid(color='w', linestyle='-', linewidth=2)

plt.show()
```
