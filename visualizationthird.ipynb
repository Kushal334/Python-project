```python
x1 = [x1 * 1 for x1 in range(0,70)]
y1 = [math.sqrt(x1) for x1 in x1]
plt.plot(x1,y1)
plt.title('$\sqrt{x}$');
```

```python
x2 = np.linspace(-5,5,100)
y2 = x2 ** 3
plt.plot(x2,y2)
plt.title('$x^{3}$');
```

```python
x3 = np.linspace(-5,5,100)
y3 = 2**x3
plt.plot(x3, y3)
plt.title('$2^{x}$');
```

```python
x1 = [x1 * 1 for x1 in range(0,70)]
y1 = [math.sqrt(x1) for x1 in x1]
x2 = np.linspace(-5,5,100)
y2 = x2 ** 3
x3 = np.linspace(-5,5,100)
y3 = 2**x3
x4 = np.linspace(-5,5,100)
y4 = 1/(x4+1)


plt.figure(figsize = (10,8))


plt.plot(x1,y1,color = 'blue', marker = '.',
         linestyle = '-', label = '1. y = √x')

plt.plot(x2,y2,color = 'r', marker = '',
         linestyle = '-', label = '2. y = x^3')

plt.plot(x3,y3,color = 'k', marker = '*',
         linestyle = '-', label = '3. y = 2^x')

plt.plot(x4,y4,color = 'm', marker = ',',
         linestyle = '-', label = '4. y = 1/(x+1)')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('4 equations')
plt.grid('True', ls = '--')
plt.legend();
```
